## Question 2

Explain the relationship between **Primary Keys** and **Foreign Keys**.

Type your answer below this line:
_______________________________________________________________________________________________________________________________
A Primary Key is a unique identifier for a table. A Foreign Key is an identifier that is unique to a table that is being joined with another-it *is* unique to table B, but table A has multiple instances of specific values.
Like: table A is employees, table B is offices, all employees have data on what office they work in. Employee Number is the primary key for A, office code is primary key for B. If joined, then office code is considered a foreign key for A.

## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
- Inner, default: only includes row if the data exists in BOTH A and B
- Left: includes all rows in table A, includes data from B that exists in A (A that is NOT in B has null values)
- Right: A and B reversed from Left Join (not in sqlite3)
- Outer: all rows in A and B included, null values if missing from either
- can also specifically *exclude* the combined (Inner Join) section, by searching for only null values
